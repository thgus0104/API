#데이터 표준화

```python
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
```

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
data1 = iris.data
data2 = scale(iris.data)
```

```python
print("old mean", np.mean(data1, axis=0)) 
#각 feature별로 평균값 제각각(범위가 다르니까)
print("new mean", np.mean(data2, axis=0))
#scale해주면 표준정규분포로 바뀌어 평균값이 비슷한 수준에 형성됨
```

```python
print("old std", np.std(data1, axis=0))
print("new std", np.std(data2, axis=0))
#새로운 표준편차는 무조건 1로 통일(정규분포니까)
```

# 데이터 정규화

```python
from sklearn.preprocessing import scale, normalize
import numpy as np
x = (np.arange(10, dtype=np.float) - 3).reshape(-1,1)
y1 = scale(x)
y2 = normalize(x)
```

```python
print("original x:\n", x)
print("scale:\n", y1)
print("norms (scale)\n", np.linalg.norm(y1, axis=1))
print("normalize:\n", y1)
print("norms (normalize)\n", np.linalg.norm(y2, axis=1))
```

# 상관계수 구하기
- -1<r<1
-양수이면 x증가>y증가
-음수이면 x증가>y감소

```python
#샘플 데이터1
from numpy import mean
from numpy import std
from numpy.random import randn
from numpy.random import seed
import matplotlib.pyplot as plt

seed(1)

data1 = 20 *randn(1000) + 100
data2 = data1 + (10*randn(1000) +50)

print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f'% (mean(data2), std(data2)))

plt.scatter(data1, data2)
```

## 방법1: 수식으로 구하기


```python
mean1= mean(data1)
std1= std(data1)
mean2= mean(data2)
std2= std(data2)

Qx = data1 - mean1
Qy = data2 - mean2
Q = Qx * Qy
```

```python
Q.shape
print(len(Q))
```

```python
r = sum(Q)/((len(Q)-1)*std1*std2)
print('correlation coefficient:',r)
```

## 방법2: pandas로 구하기

```python
import pandas as pd

rows = [int(i) for i in range(len(data1))]
s_d1 = pd.Series(data1, index=rows)
s_d2 = pd.Series(data2, index=rows)
clm = ['data1', 'data2']
d = pd.DataFrame(columns=clm)
d['data1'] = s_d1
d['data2'] = s_d2
d.corr()
```

## 방법3: Scipy의 pearson 상관계수로 구하기

```python
from scipy.stats import pearsonr

corr1, _ = pearsonr(data1, data2)

print('Pearsons correlation between data1 and data2: %.3f' % corr1)
```

## Iris data target-feauture 상관계수 구하기

```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris() #numpy array임
data_df = pd.DataFrame(iris.data, columns=iris.feature_names) #dataframe으로 변환 필요
```

```python
corr_matrix = data_df.corr()
```

```python
data_df.head()
```

```python
corr_matrix['petal width (cm)'].sort_values(ascending=False)
#petal width를 기준으로 다른 feature와의 상관계수를 한꺼번에 구할 수 있음
```

```python
data_df.columns
```

```python
#모든 feature를 기준으로 한 모든 feature와의 상관계수 한꺼번에 아는 법
from pandas.plotting import scatter_matrix
attributes = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']
scatter_matrix(data_df[attributes],figsize=(12, 8))
```
