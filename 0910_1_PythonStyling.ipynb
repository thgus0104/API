```python
import pandas as pd
import numpy as np

np.random.seed(88)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))], axis=1)
df.iloc[3, 2] = np.nan
df.iloc[0, 3] = np.nan
```

```python
print(df)
```

```python
df.style.highlight_max(color='green').highlight_min(color='pink')
```

```python
df.style.highlight_max(axis=1, color='pink')
```

```python
df.style.highlight_null(null_color='pink')

#null 값의 경우 색을 지정할때 그냥 color가 아니고 null_color로 지정 필요
```

```python
df.style.set_na_rep('OS').highlight_null(null_color='pink')
```

```python
df.style.background_gradient()
#기본컬러 PuBu
```

```python
df.style.background_gradient(cmap='binary')
#컬러차트 입력 =>cmap
```

```python
df.style.background_gradient(axis=1)
```

```python
df.style.set_properties(**{'border': '1.3px solid green', 'color': 'brown'})
```

```python
df.style.bar(color='green')
```

```python
df.style.set_precision(2)
```

```python
df.style.set_caption("This is Sample Data")
```

```python
df.style.hide_index()
```

```python
df.style.hide_columns('C')
```

```python
df.style.format('{:.3%}')
#NULL값에도 %가 붙음
```

```python
df.style.format('{:.3%}', na_rep="nan")
```
